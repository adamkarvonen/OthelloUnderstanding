```python
plt.rcParams['figure.dpi'] = 100

true_board_RR = t.randint(-1, 2, (8, 8))
recon_board_RR = t.randint(-1, 3, (8, 8))

# Create a color map with specific colors
# Creating a dictionary for the color mapping
color_map = {-1: 'red', 0: 'white', 1: 'blue', 2: 'grey'}

# Replace board values with corresponding colors using a numpy vectorized operation
label_colors = np.vectorize(color_map.get)(true_board_RR.numpy())

# Create a figure and axis for the plot
fig, ax = plt.subplots()

# Create a color map based on the unique labels in the board
unique_labels = np.unique(recon_board_RR.numpy())
colors = [color_map[label] for label in unique_labels]
cmap = plt.matplotlib.colors.ListedColormap(colors)

# plot true board rr using the true color scheme
plt.imshow(recon_board_RR, cmap=cmap)
# add circles on each square with black borders
for i in range(8):
    for j in range(8):
        circle = plt.Circle((j, i), 0.3, color=color_map[true_board_RR[i, j].item()], edgecolor='black', fill=True)
        circle_edges = plt.Circle((j, i), 0.3, color='black', fill=False)
        plt.gca().add_artist(circle)
        plt.gca().add_artist(circle_edges)
        # add grid
        plt.axhline(y=i - 0.5, color='black', linewidth=1)
        plt.axhline(y=7.5, color='black', linewidth=1)
        plt.axvline(x=i - 0.5, color='black', linewidth=1)
        plt.axvline(x=7.5, color='black', linewidth=1)
```
